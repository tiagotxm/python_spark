# Group By e Reshape

https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.groupby.html

https://pandas.pydata.org/pandas-docs/stable/user_guide/reshaping.html

```python
import numpy as np
import pandas as pd
```

```python
cidadesDigitais = pd.read_csv('cidades_digitais.csv', sep = ';', encoding = 'latin-1')
```

```python
cidadesDigitais.head()
```

```python
cidadesDigitais.head()
```

```python
cidadesDigitais.count()
```

```python
del cidadesDigitais['IBGE']
```

```python
cidadesDigitais.head()
```

```python
# Aplicando Group By
cidadesUF_1 = cidadesDigitais.groupby('UF')
```

```python
type(cidadesUF_1)
```

```python
cidadesUF_1.groups
```

```python
# Número de grupos
len(cidadesUF_1.groups)
```

```python
ordenaUF = cidadesUF_1.size()
```

```python
ordenaUF.head()
```

```python
# Ordena os grupos por ordem decrescente
ordenaUF.sort_values(ascending=False)
```

```python
cidadesUF_2 = cidadesDigitais.groupby(['UF','STATUS'])
```

```python
cidadesUF_2.groups
```

```python
# Total agrupado
cidadesUF_2.size()
```

```python
cidadesDigitais.head()
```

```python
cidadesUF_6 = cidadesUF_6.groupby(['UF','STATUS'])
```

```python
cidadesUF_6.sum()
```

```python
cidadesUF_6.aggregate(np.sum)
```
